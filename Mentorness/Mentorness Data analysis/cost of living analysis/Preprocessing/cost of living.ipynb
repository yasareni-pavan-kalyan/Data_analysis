```python
import matplotlib.pyplot as plt

# Define the categorical columns
categorical_columns = ['Avg_monthly_salary', 'Transportation', 'Childcare', 'Market', 'Clothing', 'Restaurant', 'Utilities', 'Leisure', 'Rent']

# Set the number of rows and columns for the subplots
num_rows = 3
num_cols = 3

# Create the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten the axes array
axes = axes.flatten()  # Flatten the axes array to convert it into a 1D array

# Check if 'Avg_monthly_salary' exists in categorical_data dataframe
if 'Avg_monthly_salary' not in processed.columns:
    # Remove 'Avg_monthly_salary' from categorical_columns
    categorical_columns = [col for col in categorical_columns if col != 'Avg_monthly_salary']

# Iterate over the categorical columns and plot scatterplots
for i, column in enumerate(categorical_columns):
    axes[i].scatter('basic_cost', processed[column])
    axes[i].set_xlabel('Basic Cost')
    axes[i].set_ylabel(column)
    axes[i].set_title(f'Scatterplot: Basic Cost vs {column}')

plt.tight_layout()  # Adjust the spacing between subplots
plt.show()
```

```python
# Define the base city (Country A, City X)
base_country = 'United States'
base_city = 'New York'

# Define weights for each category (example)
weights = {
    'Transportation': 0.1,
    'Childcare':0.1,
    'Clothing':0.2,
    'Restaurant': 0.1,
    'Utilities': 0.3,
    'Leisure':0.1,
    'Rent':0.1
}

# Function to calculate price index for a category relative to the base
def calculate_price_index(row, base_row):
    return (row / base_row) * 100

# Filter data for the base city
base_data = processed[(processed['country'] == base_country) & (processed['city'] == base_city)].iloc[0]

# Calculate price indices for each category relative to the base city
processed['Transportation Index'] = calculate_price_index(processed['Transportation'], base_data['Transportation'])
processed['Childcare Index'] = calculate_price_index(processed['Childcare'], base_data['Childcare'])
processed['Clothing Index'] = calculate_price_index(processed['Clothing'], base_data['Clothing'])
processed['Restaurant Index'] = calculate_price_index(processed['Restaurant'], base_data['Restaurant'])
processed['Utilites Index'] = calculate_price_index(processed['Utilities'], base_data['Utilities'])
processed['Leisure Index'] = calculate_price_index(processed['Leisure'], base_data['Leisure'])
processed['Rent Index'] = calculate_price_index(processed['Rent'], base_data['Rent'])

# Calculate the cost of living index for each city
processed['Cost of Living Index'] = (
    weights['Transportation'] * processed['Transportation Index'] +
    weights['Childcare'] * processed['Childcare Index'] +
    weights['Clothing'] * processed['Clothing Index'] + 
    weights['Restaurant'] * processed['Restaurant Index'] +
    weights['Utilities'] * processed['Utilites Index'] +
    weights['Leisure'] * processed['Leisure']+
    weights['Rent'] * processed['Rent Index']
)

# Display the cost of living index for each city
col_index_data = processed[['country', 'city', 'Cost of Living Index']]
print(col_index_data)
```

```python
# Define the base city (Country A, City X)
base_country = 'United States'
base_city = 'New York'

# Filter data for the base city
base_data = processed[(processed['country'] == base_country) & (processed['city'] == base_city)].iloc[0]

# Function to calculate the Purchasing Power Index
def calculate_purchasing_power_index(row, base_row):
    return (row['Avg_monthly_salary'] / row['Cost of Living Index']) * 100 / (base_row['Avg_monthly_salary'] / base_row['Cost of Living Index'])

# Calculate the Purchasing Power Index for each city
processed['Purchasing Power Index'] = processed.apply(calculate_purchasing_power_index, axis=1, base_row=base_data)

# Display the Purchasing Power Index for each city
ppi_data = processed[['country', 'city', 'Purchasing Power Index']]
print(ppi_data)
```
