Matches data processing

```python
matches.info()
```

```python
matches_1 = matches.copy()
```

```python
matches_1['method'].fillna('Normal', inplace = True)
matches_1['city'].fillna('Not available', inplace = True)
matches_1['result_margin'].fillna(0, inplace = True)
```

```python
matches_1.dropna(subset = ['target_overs', 'winner'], inplace = True)
```

```python
matches_1.isna().sum()
```

```python
matches_1.head()
```

```python
matches_1 = matches_1.rename(columns = {'id':'match_id'})
```

```python
matches_1.to_csv(r"C:\Users\yasar\OneDrive\Desktop\IPL Analysis\Cleaned_data\matches_cleaned.csv", index = None)
```

```python
merged = pd.merge(matches_1, deliveries_1, on = 'match_id')
```

```python
merged_clean = merged[['match_id', 'batting_team', 'bowling_team', 'batter', 'bowler', 'venue', 
        'season', 'match_type', 'batsman_runs', 'extra_runs', 'total_runs', 'is_wicket']]
```
